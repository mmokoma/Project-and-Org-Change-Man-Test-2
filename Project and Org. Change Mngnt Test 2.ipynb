```python
import requests
import lxml.html as lh
import pandas as pd
df = pd.DataFrame()
```

```python
#Reading the data
my_url = 'https://www.worldometers.info/coronavirus/#main_table'

page = requests.get(my_url) # create a handle to for contents of the wiki page

doc = lh.fromstring(page.content) # store content of the wiki page under doc

tr_elements = doc.xpath('//tr') # parse data stored between tr in the html

[len(T) for T in tr_elements[:12]] # check the length of the first 12 rows
```

```python
tr_elements = doc.xpath('//tr') # parse first row as header

col = [] # create empty list
i = 0

for t in tr_elements[0]: # for each row, store each first element (header) and an empty list
    i+=1
    name=t.text_content()
    print("%d:%s" % (i,name))
    col.append((name,[]))
```

```python
for j in range(1,len(tr_elements)): # Because header is the first row, data would be store in the subsequent rows.
    T = tr_elements[j] #T is j'th row
    
    if len(T)!=22: #if row is not size 3, //tr data is not from the table.
        break
        
    i = 0 #i is the index of the first column
    
    for t in T.iterchildren(): #iterate through each element of the row
        data=t.text_content()
            
        col[i][1].append(data) #append the data to the empty list of the i'th column
            
        i+=1 #increment i for the next column
```

```python
[len(C) for (title,C) in col]
```

```python
Dict = {title:column for (title,column) in col}
df = pd.DataFrame(Dict)
```

```python
# function to clean the dataset
def dataframeCleaner(df):
    for columnname in df: #looping through titles of the table
        temp = []
        for column in df [columnname]: #geting column elements for the each title
            column = str(column)
            column = column.replace(',','') # Removing unwanted data clutter
            column = column.replace('+','') #Removing unwanted '+'sign
            column = column.replace('\n','') #Removing unwanted '\n'
            column = column.replace('N/A','') #Removing unwanted 'N/A'
            try : #using try except block to convert datatype string to integer while avoiding error
                column = int(column)
            except: 
                pass
            temp.append(column)
        df[columnname] = temp
    df = df.drop(df.tail(1).index) # Deleting the last row
    df = df.replace(r'^\s*$',0,regex=True) # converting empty string to 0
    return df
```

<b>Conclusion:</b>

Since the p-value is < 0.001, the correlation between New Recovered cases and New Deaths is statistically significant, and the linear relationship is extremely strong (~0.976). Meaning an <b>$increase$</b> in <b>new recovered cases</b> lead to a <b>$decrease$</b> in <b>new deaths reported</b>. Therefore the variables are dependent.

```python
pearson_coef, p_value = stats.pearsonr(df['ActiveCases'], df['NewDeaths'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

<b>Simple Linear Regression</b>

$$Y = MX + C$$

```python
#Import libraries 
from sklearn.linear_model import LinearRegression
```

```python
#Using reported New Cases to predict New Deaths
x = my_df[['NewCases']] 
y = my_df['NewDeaths'] 
lm = LinearRegression() 
lm.fit(x,y)

Yhat=lm.predict(x)
Yhat[0:5]
```

```python
#Calculating the value of the intercept, c
lm.intercept_
```

```python
#Calculating the value of the slope, m
lm.coef_
```

```python
print('The value of R-squared is: ', lm.score(x, y))
```

```python
#import the function mean_squared_error from the module metrics
from sklearn.metrics import mean_squared_error
```

```python
y=lm.predict(x)
MSE = mean_squared_error(my_df['NewDeaths'], y)
print('The value of Mean Squared Error (MSE) is', MSE)
```

```python
#calculate RMSE
#import necessary libraries
from math import sqrt
```

```python
RMSE = sqrt(mean_squared_error(my_df['NewDeaths'], y))
print('The value of RMSE is' , RMSE)
```
